# MELD

Датасет с множественными модальностями, созданный на основе сериала "Друзья". Содержит разметку по эмоциям и тональностям.

## Список модальностей

*   Текст
*   Аудио
*   Видео



## Объем

*   1400 диалогов;
*   13000 высказываний (utterance);



## Особенности

*   Используются нарезанные сегменты из сериала с игрой актеров;
*   Содержит постронние шумы - закадровый смех, музыку, иногда слабо слышен голос основного говорящего;



## Аннотирование

*   Разметка эмоций по системе Пола Экмана;
*   6 базовых эмоций с добавлением нейтрального класса;
*   Разметка по 3 классам тональности - нейтральный, позитивный, негативный;



## Метрики

F1 = 2 * (precision * recall) / (precision + recall)

'weighted': каждой метке добавляется значение значимости или веса.

# OMGEmotion Recognition Dataset

*   Видео

# Вывод

**Приемущества MELD**:
* Большой объем диалогов и высказываний
* Наличие 3-х модальностей
* 7 базовых эмоций
* Имеет аннотацию тональности
* В общей сложности более 400 спикеров

**Недостатки MELD**:
* Закадровый смех и музыка
* Преобладает нейтральный класс

**Приемущества OMGEmotion**:
* Оценка анотации производится с контекстом
* Для каждого видео есть в общей сложности 5 различных аннотаций
* 6 базовых эмоций с добавлением нейтрального класса
* Подойдет для решения проблемы кластеризации

**Недостатки OMGEmotion**:
* Видеозаписи берутся из YouTube, которые имеют посторонний шум
* Не сбалансирован по разметке эмоций, значительно преобладает нейтральный класс

**Приемущества CREMA-D**:
* Большой объем записей
* Наличие 3-х модальностей
* Студийная запись как видео, так и звука
* Подробное описание датасета
* Отдельная оценка для видео, аудио и видео + аудио
* Сбалансированные распределения эмоций

**Недостатки CREMA-D**:
* Текстово ограничен 12 возможными высказываниями
* Отсутствие популярности применения
